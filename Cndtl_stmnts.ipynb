```python
## Greetings based on time of day
time = int(input("Enter the current hour (0-23): "))
if time > 6 and time < 12:
    print("Good morning!")
else :
    if time >= 12 and time < 18:
        print("Good afternoon!")
    else:
        if time >= 18 and time < 22:
            print("Good evening!")
        else:
            print("Good night!")
```
